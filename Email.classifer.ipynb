```python
df.loc[df['Category'] == 'spam', 'Category',] = 0
df.loc[df['Category'] == 'ham', 'Category',] = 1
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.Category,test_size = 0.2, random_state = 3)
```

```python
X_train
```

```python
cv = CountVectorizer()
```

```python
X_train_count = cv.fit_transform(X_train.values)
```

```python
X_train_count
```

```python
model = MultinomialNB()
model.fit(X_train_count, y_train.astype(int))
```

```python
input=[" You a spa day! Call to claim."]
input_count = cv.transform(input)
prediction = model.predict(input_count)

if prediction[0] == 1 :
    print('Ham')
else :
    print('Spam')
```
